# List

```python
list = ["apple", "banana", "guava"]

print(type(list))

for i, fruit in enumerate(list):
  print(f"Fruit no.{i+1} is {fruit}")

list.append("strawberry")

count = 0

for fruit in list:
  count+=len(fruit)

print(f"Total list characters are {count}")
```

```python
numbers = [4,6,8,23,12,78,65,2,5,1]
print(f"Original list -> {numbers}")

numbers.sort()
print(f"List after sorting -> {numbers}")
```
